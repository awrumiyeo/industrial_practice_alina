# Load the data and prepare the environment

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6) 


df = pd.read_csv('Downloads/telecom_churn.csv')

print(df.head())
```

# Churn Rate by Contract Renewal

```python
plt.figure(figsize=(8, 6))
ax = sns.countplot(x='ContractRenewal', hue='Churn', data=df)

plt.title('Churn Count by Contract Renewal Status', fontsize=16)
plt.xlabel('Contract Renewed', fontsize=12)
plt.ylabel('Number of Customers', fontsize=12)

ax.set_xticklabels(['No', 'Yes'])
plt.legend(title='Churn', labels=['No', 'Yes'])

plt.show()
```

# Churn by Customer Service Calls

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Churn', y='CustServCalls', data=df)

plt.title('Distribution of Customer Service Calls by Churn Status', fontsize=16)
plt.xlabel('Churn Status', fontsize=12)
plt.ylabel('Number of Customer Service Calls', fontsize=12)
plt.xticks([0, 1], ['Did Not Churn', 'Churned']) 

plt.show()
```

# Monthly Charge and Its Relation to Churn

```python
plt.figure(figsize=(12, 7))
sns.histplot(data=df, x='MonthlyCharge', hue='Churn', multiple='stack', kde=True, palette='viridis')

plt.title('Distribution of Monthly Charges by Churn Status', fontsize=16)
plt.xlabel('Monthly Charge ($)', fontsize=12)
plt.ylabel('Number of Customers', fontsize=12)
plt.legend(title='Churn', labels=['Yes', 'No'])

plt.show()
```

# : Scatter Plot: Monthly Charge vs. Daily Minutes

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df, x='MonthlyCharge', y='DayMins', hue='Churn', alpha=0.6, palette='magma')

plt.title('Monthly Charge vs. Total Day Minutes by Churn', fontsize=16)
plt.xlabel('Monthly Charge ($)', fontsize=12)
plt.ylabel('Total Day Minutes', fontsize=12)
plt.legend(title='Churn', labels=['Churned', 'Did Not Churn'])

plt.show()
```

# Correlation Heatmap

```python
plt.figure(figsize=(12, 10))
corr_matrix = df.corr()

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)

plt.title('Correlation Matrix of Telecom Features', fontsize=16)
plt.show()
```
